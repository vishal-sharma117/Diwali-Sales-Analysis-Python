```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('/content/Diwali Sales Data.csv', encoding='latin1')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

- Droping the column Status and unnamed1 because no value in both tha table

```python
df.drop(columns=['Status','unnamed1'],inplace=True)
```

- Now we can see both table has been droped

```python
df.info()
```

```python
df.isnull().sum()
```

- In  Amount column have some na value hence we drop that value

```python
df.dropna(inplace=True)
df.isnull().sum()
```

- in Amount Data type is float than we will change it to int

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df.info()
```

```python
df.describe()
```

- # Exploratory Data Analysis

- ### Gender

```python
ax = sns.countplot(x='Gender',data=df)
ax.bar_label(ax.containers[0])
plt.title('Sales Count According to Gender')

plt.show()

```

```python
# Ploting bar chart according to gender Vs amount
gn = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
gn.plot(kind='bar',x= 'Gender',y='Amount')
plt.title('Amount spend according to Gender')
plt.show()
```

- From the Above Graph we can see most of the buyer is female

```python
# According to Age we can check
ax= sns.countplot(x='Age Group',data=df,hue='Gender',palette='husl')
ax.bar_label(ax.containers[0])
plt.title('According To Age Group Sales')
plt.show()
```

```python
sale_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
plt.figure(figsize=(10,5))
plt.title('Total Amount Vs Age Group')
ax=sns.barplot(data=sale_age,x='Age Group',y='Amount',hue='Age Group')
plt.show()

```

- From Above Graph we can see that most of buyer are of age group between 26-35 years Female

##State

```python
sales_State =df.groupby(['State'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False)
plt.figure(figsize=(15,5))
plt.title('State VS Orders')
sns.barplot(x='State',y='Orders',data=sales_State,palette='husl')
plt.xticks(rotation=90)
plt.show()
```

- From Above graph we can se the total Number of orders which Sate has given

```python
# amount and Sales
sales_States = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False).head(10)
plt.figure(figsize=(15,5))
plt.title('Sales Amount According to Sates')
sns.barplot(data=sales_States,x='State',y='Amount',palette='husl',hue=None)
plt.show()
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


## Merital Sattus

```python
plt.figure(figsize=(8,4))
ax = sns.countplot(x='Marital_Status',data= df)
plt.title('Count Of Order Placed According to Marital_Status')
ax.bar_label(ax.containers[0])
```

```python
M_s = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(data=M_s,x='Marital_Status',y='Amount',hue='Gender',palette='husl')
plt.title('Sales Amount According to Merital_Status & Gender')

```

- Occupation

```python
plt.figure(figsize=(15,5))
ax = sns.countplot(x='Occupation',data=df,color='black')
ax.bar_label(ax.containers[0])
plt.xticks(rotation=90)
plt.show()
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

## Product category

```python
plt.figure(figsize=(15,5))
ax= sns.countplot(data=df,x='Product_Category',color='black')
plt.title('Sales Count According to Product')
ax.bar_label(ax.containers[0])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,5))
sales_product = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(data=sales_product,x='Product_Category',y='Amount',palette='viridis')
plt.title('Amount Spent Accotding to Product_Category')
plt.xticks(rotation=90)
plt.show()
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
plt.figure(figsize=(15,5))
sales_pruduct_id = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.barplot(data=sales_pruduct_id,x='Product_ID',y='Orders',palette='summer')
plt.title('Sales based on product id')
plt.show()
```

```python
# top 10 Product Sales
plt.figure(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().sort_values(ascending=False).head(10).plot(kind='bar',color='black')
plt.title('Top 10 Orders Based on Product')
```

## *Married women age group 26-35 yrs from UP,  Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category*
